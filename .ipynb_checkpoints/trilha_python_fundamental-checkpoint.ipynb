```python
# Trilha - Python - Digital Innovation One - Recofe-PE - 12/09/2024
# Desafio Fundamentos Python
# CARLOS SOARES
# https://github.com/jcarlossc

banco = ' PYBANK '
menu = f'\n{banco.center(48, "=")}\n\n[d] - DEPÓSITO\n[s] - SAQUE\n[e] - EXTRATO\n[q] - SAIR\n\n{"=" * 48}\n\n'
saldo = 0
extrato = ''
limite_por_saque = 500
numero_de_saques = 0
LIMITE_DIARIO_SAQUES = 3
valor_extrato_1 = 0
valor_extrato_2 = 0
quantidade_extrato = 0

while True:
    
    tipo_movimentacao = input(menu)
    
    if tipo_movimentacao == 'd':
        valor_deposito = float(input('Digite o valor do depósito: '))
    
        if valor_deposito > 0:
            saldo += valor_deposito
            extrato += f'Depósito: R$ {valor_deposito:.2f}\n'
            
        else:
            print('Operação falhou! O valor digitado é inválido.')  
            
    elif tipo_movimentacao == 's':
        valor_saque = float(input('Digite o valor do saque: '))
        
        if numero_de_saques < LIMITE_DIARIO_SAQUES:
            
            if valor_saque <= limite_por_saque:
                
                if valor_saque <= saldo:
                    saldo -= valor_saque
                    numero_de_saques += 1
                    extrato += f'Saque: R$ {valor_saque:.2f}\n'
                    
                else:
                    print('Operação falhou! O valor digitado é inválido.') 

            else:
                print('Operação falhou! O saque excedeu o valor de R$ 500,00.')
                
        else:
            print('Operação falhou! Número máximo de saques excedido.')
            
    elif tipo_movimentacao == 'e':
        quantidade_extrato += 1
        valor_extrato_1 += 1.89
        valor_extrato_2 = 1.89
        saldo -= valor_extrato_2
        extrato_titulo = ' EXTRATO ' 
        
        print(f'\n{extrato_titulo.center(48, "=")}')
        print('Não foram realizadas movimentações.\n\n' if not extrato else extrato)
        print(f'[{quantidade_extrato}]Extrato(s): R$ {valor_extrato_1:.2f}')
        print(f'Saldo: R$ {saldo:.2f}')
        print('=' * 48)
        
    elif tipo_movimentacao == 'q':
        break
        
    else:
        print('Operação inválida, por favor selecione novamente a operação desejada.')    

print('=' * 48)
print(f'{banco.center(48, "=")}')  
print('=' * 48)
```
